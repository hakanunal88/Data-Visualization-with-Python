```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.boxplot(data=df, x="reading score", y="parental level of education", hue="test preparation course", palette="Set2")
plt.legend(bbox_to_anchor=(1,0.6))
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.violinplot(data=df, x="reading score", y="parental level of education", bw = 0.2) #split=True, inner="stick"
#plt.legend(bbox_to_anchor=(1.2,0.6))
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x="math score",y="gender", size=2, hue="test preparation course", dodge=True)
plt.legend(bbox_to_anchor=(1.3,0.6))
```

```python
sns.boxenplot(x="math score", y="test preparation course", data=df, hue="gender")
```

```python
sns.jointplot(data=df, x="math score", y="reading score", kind="scatter", hue="gender")
```

```python
sns.pairplot(data=df, hue="gender", corner=True) #diag_kind="hist"
```
