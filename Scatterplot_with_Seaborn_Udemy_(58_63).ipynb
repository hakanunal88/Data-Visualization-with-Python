```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education", palette="Dark2", style="level of education", s=200, alpha=1);
```

```python
#plt.figure(figsize=(5,8), dpi=200)
sns.rugplot(x="salary", data=df, height=0.5);
```

```python
#sns.set(style="darkgrid")
sns.displot(data=df, x="salary", kde=True, rug=True); # bins=20, color="red", edgecolor="blue", linewidth=4, ls="--"
```

```python
sns.histplot(data=df, x="salary");
```

```python
sns.kdeplot(data=df, x="salary");
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=["age"])
```

```python
sns.displot(data=sample_ages, x="age", rug=True, bins=30, kde=True);
```

```python
sns.kdeplot(data=sample_ages, x="age", clip=[0,100], bw_adjust=0.6, shade=True);
```

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x="level of education", hue="division", palette="Set2");
#plt.ylim(0,1000)
```

```python
df["level of education"].value_counts()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x="level of education", y="salary", estimator=np.mean, ci="sd", hue="division")
plt.legend(bbox_to_anchor=(1,1));
```
